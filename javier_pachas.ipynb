```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Uploading datasets
df_a =  pd.read_csv('/content/task_a.csv')
df_b =  pd.read_csv('/content/task_b.csv')
df_countries =  pd.read_csv('/content/countries.csv')
```

```python
df_a
```

```python
# Convert 'Month ID (YYYYMM)' to datetime
df_a['Month ID (YYYYMM)'] = pd.to_datetime(df_a['Month ID (YYYYMM)'], format='%Y%m')
```

## Task A1

```python
coefficients = np.polyfit(range(1,25), df_a['Defects'], 1)
slope, intercept = coefficients

line1 = slope * range(1,25) + intercept
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_a['Month ID (YYYYMM)'], df_a['Defects'], marker='o')
plt.plot(df_a['Month ID (YYYYMM)'], line1, color='red', label=f'Fitted Line: y = {slope:.2f}x + {intercept:.2f}')
plt.title('Task A1: Defects Over Time')
plt.xlabel('Month')
plt.ylabel('Defects')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```

##Task A2

```python
coefficients = np.polyfit(range(1,25), df_a['Opportunities'], 1)
slope, intercept = coefficients

line2 = slope * range(1,25) + intercept
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_a['Month ID (YYYYMM)'], df_a['Opportunities'], marker='o', color='orange')
plt.plot(df_a['Month ID (YYYYMM)'], line2, color='blue', label=f'Fitted Line: y = {slope:.2f}x + {intercept:.2f}')
plt.title('Task A2: Opportunities Over Time')
plt.xlabel('Month')
plt.ylabel('Opportunities')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend()
plt.show()
```

##Task A3

```python
plt.figure(figsize=(10, 6))
plt.plot(df_a['Month ID (YYYYMM)'], df_a['Defect Rates'], marker='o', label='Defect Rates')
plt.plot(df_a['Month ID (YYYYMM)'], df_a['Mean Rate'], linestyle='--', color='red', label='Mean Rate')
plt.plot(df_a['Month ID (YYYYMM)'], df_a['2 Sigma Limit'], linestyle='--', color='green', label='2 Sigma Limit')
plt.plot(df_a['Month ID (YYYYMM)'], df_a['3 Sigma Limit'], linestyle='--', color='blue', label='3 Sigma Limit')
plt.title('Task A3: Defect Rates with Mean Rate and Limits Over Time')
plt.xlabel('Month')
plt.ylabel('Rate')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

##Task B1

```python
df_b['Date'] = pd.to_datetime(df_b['Date'], format='%d/%m/%y')

df_merged = pd.merge(df_b, df_countries, on='Account.Number', how='left')

#First 5 rows of the merger dataframe
df_merged.head()
```

## Task B2

```python
df_merged['Year'] = df_merged['Date'].dt.year
product_summary = df_merged.groupby(['Product', 'Year'])['Quantity'].sum().unstack(fill_value=0)
product_summary['Percent Difference'] = ((product_summary[2021] - product_summary[2020]) / product_summary[2020]) * 100
```

```python
product_summary
```

##Task B3

```python
complaints_by_country = df_merged[df_merged['Date'].dt.year == 2020].groupby('Country')['Complaint.Number'].count()
top_3_countries = complaints_by_country.sort_values(ascending=False).head(3)
```

```python
print("Top 3 Countries in 2020:\n", top_3_countries)
```

##Task B4

```python
df_merged['defect_category'] = df_merged['Defect'].str.split('-').str[0]
complaints_by_defect = df_merged.groupby('defect_category')['Complaint.Number'].count()
```

```python
print("\nComplaint Volumes by defect category:\n", complaints_by_defect)
```
